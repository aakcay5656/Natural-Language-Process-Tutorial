```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt
```

# İmport Data

```python
data = pd.read_csv(r"/kaggle/input/twitter-user-gender-classification/gender-classifier-DFE-791531.csv",encoding = "latin1")
```

```python
data.head()
```

## I get the necessary columns for our tutorial

```python
data = pd.concat([data.gender,data.description],axis=1)
data.dropna(axis = 0,inplace = True)
```

```python
data
```

### it is necessary to convert the gender column to numeric expression to use the library

```python
data.gender = [1 if each == "female" else 0 for each in data.gender]
```

```python
data
```

# Cleaning Data

```python
import re
import nltk as nlp # natural language tool kit
    
from nltk.corpus import stopwords 
from nltk.stem import WordNetLemmatizer
lemma=WordNetLemmatizer()
```

```python
description_list=[]
for description1 in data.description:
    
    description1=re.sub("[^a-zA-Z]"," ",description1) 
    description1=description1.lower()
    description1=nlp.word_tokenize(description1)
    description1=[word for word in description1 if not word in set (stopwords.words("english"))]

    
    #description1=[lemma.lemmatize(word) for word in description1]
    description1=" ".join(description1)
    description_list.append(description1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
max_features = 500

count_vectorizer = CountVectorizer(max_features=max_features,stop_words="english") 

sparce_matrix=count_vectorizer.fit_transform(description_list).toarray()

print("most frequently used {} words :{}".format(max_features,count_vectorizer.get_feature_names_out()))
```

## Data Preprocessing

```python
x = sparce_matrix
y = data.iloc[:,0].values

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 3)
```

# Naive Bayes Classification

```python
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train, y_train)

y_predict = nb.predict(x_test)
print("accuracy: ", accuracy_score(y_test, y_predict)*100)
```

# K-Nearest Neighbors Classification

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 13)# n_neighbors = k

knn.fit(x_train,y_train)

prediction = knn.predict(x_test)


print("{} knn score: {}".format("13",knn.score(x_test, y_test)*100))
```

# Support Vector Machines Classification

```python
from sklearn.svm import SVC
 
svm = SVC(random_state = 1)
svm.fit(x_train,y_train)

print(" accuracy of svm algo: ",svm.score(x_test,y_test)*100)
```
